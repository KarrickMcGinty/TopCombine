```python
#Change all bins with values of 0 to values of .01
import ROOT

# Open the ROOT file
file = ROOT.TFile.Open("2016ULpreVFPRebinCopy2.root", "UPDATE")

# Get the list of keys (histogram names) in the file
keys = file.GetListOfKeys()

# Iterate through each histogram
for key in keys:
    # Get the histogram by its name
    histogram = file.Get(key.GetName())
    #print(key)
    # Loop over each bin in the histogram
    for i in range(1, histogram.GetNbinsX() + 1):
        # Check if the bin value is 0
        print(histogram.GetBinContent(i))
        if histogram.GetBinContent(i) < 1:
            # Set the bin value to 1
            histogram.SetBinContent(i, 1)

    # Write the modified histogram back to the file
    histogram.Write()

# Close the ROOT file
file.Close()
```

```python
import ROOT

# Define the function to delete bins from a histogram
def delete_bins(hist, n_bins_to_delete):
    for i in range(1, n_bins_to_delete + 1):
        hist.SetBinContent(i, 0)
        hist.SetBinError(i, 0)

# Open the ROOT file
file_path = "2018UL-Copy4.root"  # Replace with the path to your ROOT file
file = ROOT.TFile.Open(file_path, "UPDATE")

# Loop through all histograms in the file
for key in file.GetListOfKeys():
    obj = key.ReadObj()

    # Check if the object is a histogram (TH1)
    if isinstance(obj, ROOT.TH1):
        print(f"Processing histogram: {obj.GetName()}")

        # Delete the first 4 bins (adjust the number as needed)
        delete_bins(obj, 4)

        # Write the modified histogram back to the file
        obj.Write(obj.GetName(), ROOT.TObject.kOverwrite)

# Close the ROOT file
file.Close()


```

```python
#SET MAXIMUM OR MINIUM BIN VALUES
import ROOT

# Open the original ROOT file for reading
input_file = ROOT.TFile.Open("2018UL_NoFirst10Bins.root", "READ")

# Create a new ROOT file for writing
output_file = ROOT.TFile.Open("2018UL_NoFirst10Bins_Minimum.root", "RECREATE")

# Get the list of keys (histogram names) in the input file
keys = input_file.GetListOfKeys()

# Iterate through each histogram
for key in keys:
    # Get the histogram by its name from the input file
    histogram = input_file.Get(key.GetName())

    # Create a copy of the original histogram
    modified_histogram = histogram.Clone()

    # Loop over each bin in the copy
    for i in range(1, modified_histogram.GetNbinsX() + 1):
        # Check if the bin value is less than 1
        if modified_histogram.GetBinContent(i) < 0.1:
            # Set the bin value to 1
            modified_histogram.SetBinContent(i, 0.1)

    # Write the modified copy of the histogram to the output file
    output_file.WriteTObject(modified_histogram, key.GetName())

# Close both the input and output ROOT files
input_file.Close()
output_file.Close()
```

```python
#RECREATE ROOT FILE WITHOUT CERTAIN BINS
import ROOT

# Open the input ROOT file
input_file = ROOT.TFile.Open("2018UL.root", "READ")

# Create a new output ROOT file
output_file = ROOT.TFile.Open("2018UL_NoFirst10Bins.root", "RECREATE")

# Get the list of keys (histogram names) in the input file
keys = input_file.GetListOfKeys()

# Loop through each key (histogram name)
for key in keys:
    # Get the histogram by its name
    histogram = input_file.Get(key.GetName())

    # Create a new histogram with the same properties (name, title, and binning)
    new_histogram = ROOT.TH1F(histogram.GetName(), histogram.GetTitle(), histogram.GetNbinsX() - 10, histogram.GetXaxis().GetXmin() + 4 * histogram.GetXaxis().GetBinWidth(1), histogram.GetXaxis().GetXmax())

    # Copy the contents from the original histogram, excluding the first 4 bins
    for i in range(10+1, histogram.GetNbinsX() + 1):
        new_histogram.SetBinContent(i - 10, histogram.GetBinContent(i))

    # Write the new histogram to the output file
    new_histogram.Write()

# Close both input and output files
input_file.Close()
output_file.Close()



```

```python
def renameHistogram(fileName, OldName, NewName):
    f = ROOT.TFile.Open(fileName, "UPDATE")
    h = f.Get(OldName)
    #h = TH1F(h) # Add this line to cast the object to a TH1F object

    # Rename the histogram
    h.SetName(NewName)

    # Delete the old histogram
    f.Delete(OldName)

    # Write the changes to the file
    f.Write("", ROOT.TObject.kOverwrite)
    f.Close()
```
