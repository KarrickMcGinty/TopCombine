```python
# Function to calculate the envelope of histograms  NEW
def envelope_histograms(histograms, ttbar_hist, ttbar_bin_value):
    num_bins = histograms[0].GetNbinsX()
    result_histogram = ttbar_hist.Clone()

    for i in range(1, num_bins + 1):
        ttbar_bin_value = ttbar_hist.GetBinContent(i)
        bin_content_list = [hist.GetBinContent(i) for hist in histograms]
        average_bin_content = sum(bin_content_list) / len(bin_content_list)

        max_bin_content = max(bin_content_list + [ttbar_bin_value])
        min_bin_content = min(bin_content_list + [ttbar_bin_value])

        if average_bin_content > ttbar_bin_value:
            result_histogram.SetBinContent(i, max_bin_content)
        elif average_bin_content < ttbar_bin_value:
            result_histogram.SetBinContent(i, min_bin_content)

    return result_histogram

def add_histograms_in_quadrature(histograms, ttbar_hist, ttbar_bin_value):
    num_bins = histograms[0].GetNbinsX()
    result_histogram = ttbar_hist.Clone()

    for i in range(1, num_bins + 1):
        ttbar_bin_value = ttbar_hist.GetBinContent(i)
        deviations = [hist.GetBinContent(i) - ttbar_bin_value for hist in histograms]
        average_deviation = sum(deviations) / len(deviations)

        sum_of_squares = sum([(deviation) ** 2 for deviation in deviations])
        quad_sum = math.sqrt(sum_of_squares)

        if average_deviation > 0:
            result_histogram.SetBinContent(i, ttbar_bin_value + quad_sum)
        else:
            result_histogram.SetBinContent(i, ttbar_bin_value - quad_sum)

    return result_histogram

# Open the root file
file = ROOT.TFile.Open(fileEnding + ".root", "UPDATE")

# Get the "ttbar" histogram
ttbar_hist = file.Get("ttbar;1")
ttbar_bin_value = ttbar_hist.GetBinContent(1)

# Envelope systematics
for syst_index, syst_name in enumerate(env_systematics):
    histograms_down = [file.Get("ttbar_" + syst + "Down;1") for syst in syst_name]
    histograms_up = [file.Get("ttbar_" + syst + "Up;1") for syst in syst_name]

    # Create envelope histograms
    envelope_hist_down = envelope_histograms(histograms_down, ttbar_hist, ttbar_bin_value)
    envelope_hist_up = envelope_histograms(histograms_up, ttbar_hist, ttbar_bin_value)

    # Save the envelope histograms in the file with ttbar_ prefix
    file.WriteTObject(envelope_hist_down, "ttbar_" + env_syst_names[syst_index] + "Down")
    file.WriteTObject(envelope_hist_up, "ttbar_" + env_syst_names[syst_index] + "Up")
    
# Envelope Symmetric systematics
for syst_index, syst_name in enumerate(symmetric_sytematics):
    histograms_down = [file.Get("ttbar_" + syst + "Down;1") for syst in syst_name]
    histograms_up = [file.Get("ttbar_" + syst + "Up;1") for syst in syst_name]

    # Create envelope histograms
    envelope_hist_down = envelope_histograms(histograms_down, ttbar_hist, ttbar_bin_value)
    envelope_hist_up = envelope_histograms(histograms_up, ttbar_hist, ttbar_bin_value)

    # Save the envelope histograms in the file with ttbar_ prefix
    file.WriteTObject(envelope_hist_down, "ttbar_" + symmetric_names[syst_index] + "Down")
    file.WriteTObject(envelope_hist_up, "ttbar_" + symmetric_names[syst_index] + "Up")


# Quadrature systematics
for syst_index, syst_name in enumerate(quad_systematics):
    histograms_down = [file.Get("ttbar_" + syst + "Down;1") for syst in syst_name]
    histograms_up = [file.Get("ttbar_" + syst + "Up;1") for syst in syst_name]

    # Create histograms with quadrature
    quad_hist_down = add_histograms_in_quadrature(histograms_down, ttbar_hist, ttbar_bin_value)
    quad_hist_up = add_histograms_in_quadrature(histograms_up, ttbar_hist, ttbar_bin_value)

    # Save the quadrature histograms in the file with ttbar_ prefix
    file.WriteTObject(quad_hist_down, "ttbar_" + quad_syst_names[syst_index] + "Down")
    file.WriteTObject(quad_hist_up, "ttbar_" + quad_syst_names[syst_index] + "Up")

# Close the root file
file.Close()
```

```python
#Envelope and Quadrature with deviation error fix
# The lists "env_systematics", "quad_systematics", "env_syst_names", and "quad_syst_names" must be defined properly before running this code.

# Function to calculate the envelope of histograms
def envelope_histograms(histograms, ttbar_hist):
    num_bins = histograms[0].GetNbinsX()
    result_histogram = ttbar_hist.Clone()

    for i in range(1, num_bins + 1):
        bin_content_list = [hist.GetBinContent(i) for hist in histograms]
        max_bin_content = max(bin_content_list + [ttbar_hist.GetBinContent(i)])
        min_bin_content = min(bin_content_list + [ttbar_hist.GetBinContent(i)])
        ttbar_bin_content = ttbar_hist.GetBinContent(i)

        if abs(max_bin_content - ttbar_bin_value) > abs(min_bin_content - ttbar_bin_value):
            result_histogram.SetBinContent(i, max_bin_content)
        else:
            result_histogram.SetBinContent(i, min_bin_content)

    return result_histogram

# Function to add histograms in quadrature
def add_histograms_in_quadrature(histograms, ttbar_hist):
    num_bins = histograms[0].GetNbinsX()
    result_histogram = ttbar_hist.Clone()

    for i in range(1, num_bins + 1):
        sum_of_squares = sum([(hist.GetBinContent(i) - ttbar_hist.GetBinContent(i)) ** 2 for hist in histograms])
        quad_sum = math.sqrt(sum_of_squares)
        ttbar_bin_content = ttbar_hist.GetBinContent(i)

        if ttbar_bin_content - quad_sum < 0:
            result_histogram.SetBinContent(i, 0)
        else:
            result_histogram.SetBinContent(i, ttbar_bin_content - quad_sum)

    return result_histogram

# Open the root file
file = ROOT.TFile.Open(fileEnding + ".root", "UPDATE")

# Envelope systematics
for syst_index, syst_name in enumerate(env_systematics):
    histograms_down = [file.Get("ttbar_" + syst + "Down;1") for syst in syst_name]
    histograms_up = [file.Get("ttbar_" + syst + "Up;1") for syst in syst_name]

    # Get the ttbar histogram
    ttbar_hist = file.Get("ttbar;1")

    # Create envelope histograms
    envelope_hist_down = envelope_histograms(histograms_down, ttbar_hist)
    envelope_hist_up = envelope_histograms(histograms_up, ttbar_hist)

    # Save the envelope histograms in the file with ttbar_ prefix
    file.WriteTObject(envelope_hist_down, "ttbar_" + env_syst_names[syst_index] + "Down")
    file.WriteTObject(envelope_hist_up, "ttbar_" + env_syst_names[syst_index] + "Up")

# Quadrature systematics
for syst_index, syst_name in enumerate(quad_systematics):
    histograms_down = [file.Get("ttbar_" + syst + "Down;1") for syst in syst_name]
    histograms_up = [file.Get("ttbar_" + syst + "Up;1") for syst in syst_name]

    # Get the ttbar histogram
    ttbar_hist = file.Get("ttbar;1")

    # Create histograms with quadrature
    quad_hist_down = add_histograms_in_quadrature(histograms_down, ttbar_hist)
    quad_hist_up = add_histograms_in_quadrature(histograms_up, ttbar_hist)

    # Save the quadrature histograms in the file with ttbar_ prefix
    file.WriteTObject(quad_hist_down, "ttbar_" + quad_syst_names[syst_index] + "Down")
    file.WriteTObject(quad_hist_up, "ttbar_" + quad_syst_names[syst_index] + "Up")

# Close the root file
file.Close()
```

```python
# Function to calculate the maximum deviation of a histogram from a reference histogram
def calculate_max_deviation(hist, ref_hist):
    num_bins = hist.GetNbinsX()
    max_deviation = 0

    for i in range(1, num_bins + 1):
        deviation = hist.GetBinContent(i) - ref_hist.GetBinContent(i)
        if abs(deviation) > max_deviation:
            max_deviation = abs(deviation)

    return max_deviation

# Open the root file
file = ROOT.TFile.Open("Total.root")

# Get the "ttbar" histogram
ttbar_hist = file.Get("ttbar;1")

# Verify enveloping
for syst_name in env_syst_names:
    hist_down = file.Get("ttbar_" + syst_name + "Down")
    hist_up = file.Get("ttbar_" + syst_name + "Up")

    # Print the values for the enveloped histograms
    print(f"Envelope verification for {syst_name}:")
    print("Bin\tttbar\tDown\tUp")
    for i in range(1, ttbar_hist.GetNbinsX() + 1):
        ttbar_value = ttbar_hist.GetBinContent(i)
        down_value = hist_down.GetBinContent(i)
        up_value = hist_up.GetBinContent(i)
        print(f"{i}\t{ttbar_value:.4f}\t{down_value:.4f}\t{up_value:.4f}")

        # Determine which direction to envelop the deviation
        if abs(ttbar_value - down_value) < abs(ttbar_value - up_value):
            hist_down.SetBinContent(i, down_value)
            hist_up.SetBinContent(i, ttbar_value + (ttbar_value - down_value))
        else:
            hist_down.SetBinContent(i, ttbar_value - (up_value - ttbar_value))
            hist_up.SetBinContent(i, up_value)

    # Calculate max deviation for the enveloped histograms
    max_deviation_down = calculate_max_deviation(hist_down, ttbar_hist)
    max_deviation_up = calculate_max_deviation(hist_up, ttbar_hist)

    print(f"Maximum deviation for {syst_name} Down: {max_deviation_down:.4f}")
    print(f"Maximum deviation for {syst_name} Up: {max_deviation_up:.4f}")
    print()

# Verify quadrature
for syst_name in quad_syst_names:
    hist_down = file.Get("ttbar_" + syst_name + "Down")
    hist_up = file.Get("ttbar_" + syst_name + "Up")

    # Print the values for the quadratured histograms
    print(f"Quadrature verification for {syst_name}:")
    print("Bin\tttbar\tDown\tUp")
    for i in range(1, ttbar_hist.GetNbinsX() + 1):
        ttbar_value = ttbar_hist.GetBinContent(i)
        down_value = hist_down.GetBinContent(i)
        up_value = hist_up.GetBinContent(i)
        print(f"{i}\t{ttbar_value:.4f}\t{down_value:.4f}\t{up_value:.4f}")

        # Calculate the quadratured deviation
        new_value_down = ttbar_value - math.sqrt(abs(ttbar_value - down_value) ** 2)
        new_value_up = ttbar_value + math.sqrt(abs(ttbar_value - up_value) ** 2)
        hist_down.SetBinContent(i, new_value_down)
        hist_up.SetBinContent(i, new_value_up)

    # Calculate max deviation for the quadratured histograms
    max_deviation_down = calculate_max_deviation(hist_down, ttbar_hist)
    max_deviation_up = calculate_max_deviation(hist_up, ttbar_hist)

    print(f"Maximum deviation for {syst_name} Down: {max_deviation_down:.4f}")
    print(f"Maximum deviation for {syst_name} Up: {max_deviation_up:.4f}")
    print()

# Close the root file
file.Close()
```

```python
def renameHistogram(file):
    
    f = ROOT.TFile.Open(file, "UPDATE")
    histogram = uproot.open(file)
    histogram.keys()
    name_old = histogram.keys()[0]
    
    # Remove the _combined_Histogram;1 at the end
    name = name_old.replace('_combined_Histogram;1', '')

    # Add ttbar_ prefix for background and background names for everything else
    if 'Nominal_combined_toponium' in name:
        name = 'signal'
    elif 'Nominal_combined_ttbars' in name:
        name = 'ttbar'
    elif 'wwtoall' in name or 'wztoall' in name or "zztoall" in name:
        name = 'diboson'
    elif 'singletop' in name or 'singleantitop' in name:
        name = 'top'
    elif 'Nominal_combined_dy' in name:
        name = 'zjets'
    else:
        name = 'ttbar_' + name

    # Replace _DOWN with Down and _UP with Up
    name = name.replace('_DOWN', 'Down')
    name = name.replace('_UP', 'Up')

    # Remove spaces
    name = name.replace(' ', '')
    name = name.replace(';1', '')
    print(name_old)
    h = f.Get(name_old)
    #h = TH1F(h) # Add this line to cast the object to a TH1F object

    # Rename the histogram
    h.SetName(name)

    # Delete the old histogram
    f.Delete(name_old)
    f.Write("", ROOT.TObject.kOverwrite)
    ROOT.TFile.Close(f)
```

```python
# ENVELOPE

def Envelope(histograms, file, output_hist):


    h = file.Get("signal;1")
    h.SetName(output_hist)
    # Get the number of entries in a specific bin (e.g., bin number 3)
    for i in range(1,21):
        new_entries = 0.0
        bin_number = i
        for histogram in histograms:
            if bin_entries > new_entries:
                new_entries = bin_entries
        h.SetBinContent(bin_number, new_entries)
```
