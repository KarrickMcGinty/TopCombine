```python
#Change all bins with values of 0 to values of .01
import ROOT

# Open the ROOT file
file = ROOT.TFile.Open("Total.root", "UPDATE")

# Get the list of keys (histogram names) in the file
keys = file.GetListOfKeys()

# Iterate through each histogram
for key in keys:
    # Get the histogram by its name
    histogram = file.Get(key.GetName())

    # Loop over each bin in the histogram
    for i in range(1, histogram.GetNbinsX() + 1):
        # Check if the bin value is 0
        if histogram.GetBinContent(i) <= 0:
            # Set the bin value to 1
            histogram.SetBinContent(i, 0.01)

    # Write the modified histogram back to the file
    histogram.Write()

# Close the ROOT file
file.Close()
```

```python
import uproot
import ROOT

def scale_signal_histogram(input_file, output_file, scaling_factor):
    # Open the input ROOT file
    root_file = uproot.open(input_file)

    # Get the "signal;1" histogram
    signal_hist = root_file["signal;1"]

    # Get the bin contents and errors
    bin_contents = signal_hist.values()
    bin_errors = signal_hist.variances()

    # Scale the bin contents by the scaling factor
    scaled_contents = bin_contents * scaling_factor

    # Create a new ROOT file for writing
    output_root_file = ROOT.TFile.Open(output_file, "RECREATE")

    # Create a clone of the original histogram with scaled bin contents
    num_bins = signal_hist.numbins
    x_min = signal_hist.xmin
    x_max = signal_hist.xmax
    scaled_hist = ROOT.TH1D("signal", "signal", num_bins, x_min, x_max)
    scaled_hist.Sumw2()

    for i in range(1, num_bins + 1):
        scaled_hist.SetBinContent(i, scaled_contents[i - 1])
        scaled_hist.SetBinError(i, bin_errors[i - 1])

    # Write the scaled histogram to the output ROOT file
    output_root_file.WriteTObject(scaled_hist, "signal;1")

    # Loop over all other keys (histograms) in the input file and write them to the output file
    for key in root_file.keys():
        if key != "signal;1":
            obj = root_file[key]
            output_root_file.WriteTObject(obj, key)

    # Close the input and output ROOT files
    root_file.close()
    output_root_file.Close()

    print(f"Scaled histogram saved to: {output_file}")

# Usage example
input_file = "2018UL.root"
output_file = "2018ULScaled.root"
scaling_factor = 3.0
scale_signal_histogram(input_file, output_file, scaling_factor)
```

```python
import numpy as np

def compare_histograms(file1, file2):
    # Open the ROOT files using uproot
    root_file1 = uproot.open(file1)
    root_file2 = uproot.open(file2)

    # Get the list of histogram names from the first ROOT file
    hist_names = root_file1.keys()

    for hist_name in hist_names:
        # Check if the histogram exists in both ROOT files
        if hist_name in root_file2:
            # Get the histograms from both ROOT files
            hist1 = root_file1[hist_name]
            hist2 = root_file2[hist_name]

            # Convert the histograms to NumPy arrays for comparison
            hist1_np = hist1.to_numpy()
            hist2_np = hist2.to_numpy()

            # Compare the histograms
            if not np.array_equal(hist1_np, hist2_np):
                print(f"Histogram {hist_name} has different bin contents.")
        else:
            print(f"Histogram {hist_name} not found in {file2}.")

    # Close the ROOT files
    root_file1.close()
    root_file2.close()

# Usage example
file1 = "Total.root"
file2 = "ManualTotal.root"
compare_histograms(file1, file2)
```

```python
def add_histograms_in_quadrature(histogram1, histogram2):
    # Get the bin contents and errors of the histograms
    bin_contents1 = histogram1.contents
    bin_errors1 = histogram1.errors
    bin_contents2 = histogram2.contents
    bin_errors2 = histogram2.errors

    # Create a new histogram to store the result
    num_bins = len(bin_contents1)
    result_histogram = ROOT.TH1D("result", "Result Histogram", num_bins, 0, num_bins)

    # Add the histograms in quadrature and store the result in the new histogram
    for i in range(1, num_bins + 1):
        bin_content1 = bin_contents1[i - 1]
        bin_error1 = bin_errors1[i - 1]
        bin_content2 = bin_contents2[i - 1]
        bin_error2 = bin_errors2[i - 1]

        # Add the bins in quadrature
        result_bin_content = (bin_content1**2 + bin_content2**2)**0.5
        result_bin_error = (bin_error1**2 + bin_error2**2)**0.5

        result_histogram.SetBinContent(i, result_bin_content)
        result_histogram.SetBinError(i, result_bin_error)

    return result_histogram

# Example usage
file1 = uproot.open("Total.root")
histogram1 = file1["signal;1"]

file2 = uproot.open("Total.root")
histogram2 = file2["signal;1"]

result_histogram = add_histograms_in_quadrature(histogram1, histogram2)

# Save the result in a new ROOT file
output_file = ROOT.TFile("result.root", "RECREATE")
result_histogram.Write()
output_file.Close()
```
